```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
#Importing necessary libraries
import requests
# !pip install bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
df=pd.DataFrame(columns=['book_name','author_name','user_rating','book_link'])
bookDF=pd.DataFrame(columns=['Publisher', 'Language', 'Paperback',' ISBN-10', 'ISBN-13', 'Weight', 'Dimensions', 'First Three Customer Reviews'])
print(len(df))
```

```python
def get_data(pageno):
    url='https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_'+str(pageno)+'_books?ie=UTF8&pg='+str(pageno)
    r=requests.get(url)
    soup=BeautifulSoup(r.content,'html.parser')
    
    for d in soup.find_all('div',attrs={'class':"zg-grid-general-faceout"}):
        book_name=d.find('div',attrs={'class':"_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y"})
        author_name=d.find('div',attrs={'class':"a-row a-size-small"})
        user_rating=d.find('span',attrs={'class':"a-icon-alt"})
        book_link=d.find('a',attrs={'class':"a-link-normal"})
        all1=[]
        if book_name is not None:
            all1.append(book_name.text)
        else:
            all1.append('NaN')
        if author_name is not None:
            all1.append(author_name.text)
        else:
            all1.append('NaN')
        if user_rating is not None:
            all1.append(user_rating.text)
        else:
            all1.append('NaN')
        if book_link is not None:
            all1.append('https://www.amazon.in'+book_link.get('href'))
        else:
            all1.append('NaN')
        #print(all1)
        df.loc[len(df)] = all1
    return df
```

```python
service = Service(executable_path=ChromeDriverManager().install())
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)
```

```python
def getBookDetails():
    bookLinks = df['book_link']
    num=1
    for url in bookLinks:
        
        driver.get(url)
        details=driver.find_element(By.ID, 'detailBullets_feature_div')
        list=details.find_elements(By.TAG_NAME,"li")

        all=[]
        for i in list[0:7]:
            a=i.text.partition(":")
            if (a[-1] is not None):
                all.append(a[-1])
            else:
                all.append("NaN")
        if (len(list)!=7):
            for i in range(7-len(list)):
                all.append('NaN')


        commentsList=driver.find_elements(By.CLASS_NAME,'a-expander-content')
        com=[]
        numberOfComments=3
        for i in commentsList[6:6+numberOfComments]:

            com.append(f"{i.text}\n--------------------")
  
        comments="\n".join(com)
        all.append(comments)
        print(f"Name: {df['book_name'][num-1]}  {num}th book out of {len(bookLinks)} done ")
        print(len(all))
        bookDF.loc[len(bookDF.index)]=all
        num=num+1
        
    return bookDF
```
